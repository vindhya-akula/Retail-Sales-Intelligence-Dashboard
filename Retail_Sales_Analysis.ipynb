# Key Performance Indicators (KPIs)

This section calculates the important business metrics used to evaluate the overall performance of the retail business.

## KPI 1 – Total Sales

### Business Purpose
Measures the total revenue generated by the business during the analysis period.

### Formula
Total Sales = Sum of Sales

### Python Function
sum()

```python
total_sales = master_df["Sales"].sum()
print("Total Sales:", total_sales)
```

# KPI 2 - Total Profit

#### Business Purpose:
Measures the total profit generated by business during the analysis

### Formula
Total Profit = Sum of Profit

### Python Function
sum()


```python
total_profit = master_df["Profit"].sum()
print("Total Profit :", total_profit)
```

# KPI 3 - Total Orders

#### Business Purpose:
Measures the total orders generated by business during the analysis

### Formula
Total Orders = Count of Order_IDs

### Python Function
Count()

```python
total_orders = master_df["Order_ID"].count()
print("Total Orders:", total_orders)
```

# KPI 4 - Total Customers

#### Business Purpose:
Measures the total customers generated by business during the analysis

### Formula
Total Customers = Number of Customers per order

### Python Function
nunique()


```python
total_customers = master_df["Customer_ID"].nunique()
print("Total Customers:", total_customers)
```

# KPI 5 - Average Rating

#### Business Purpose:
Measures the average rating generated by business during the analysis

### Formula
Average Rating = Average of Rating

### Python Function
mean()

```python
average_rating = master_df["Rating"].mean()
print("Average Rating:", average_rating)
```

## KPI 6 – Average Order Value

### Business Purpose
Measures the average amount spent by customers in each order. It helps management understand customer spending behavior and identify opportunities to increase order value.

### Formula
Average Order Value = Total Sales ÷ Total Orders

### Python Function
sum() and count()

```python
average_order_value = total_sales / total_orders
print("Average Order Value:", average_order_value)
```

## KPI 7 – Profit Margin

### Business Purpose
Measures the percentage of profit earned from total sales revenue and helps evaluate overall business profitability.

### Formula
Profit Margin (%) = (Total Profit ÷ Total Sales) × 100

### Python Function
sum()

```python
profit_margin = (total_profit / total_sales) * 100
print("Profit Margin (%):", profit_margin)
```

# Exploratory Data Analysis (EDA)

This section explores different aspects of the retail business to identify trends, patterns, and business opportunities.

## Sales by Category

### Business Question

Which product category generates the highest sales?

```python
category_sales = (
    master_df
    .groupby("Category")
    .agg(
        Total_Sales=("Sales", "sum")
    )
    .sort_values("Total_Sales", ascending=False)
)

category_sales
```

```python
category_sales.plot(
    kind="bar",
    y="Total_Sales",
    figsize=(8,5),
    title="Total Sales by Category"
)

plt.ylabel("Sales")
plt.show()
```

## Analysis 2: Total Sales by Region
### Business Question

Which region generates the highest total sales?

```python
region_sales = (
    master_df
    .groupby("Region")
    .agg(
        Total_Sales = ("Sales", "sum")
        )
    .sort_values("Total_Sales",ascending = False)
)

region_sales
```

```python
region_sales.plot(
    kind = "bar",
    y="Total_Sales",
    figsize=(8,5),
    title="Total Sales by Region"

)
plt.ylabel("Total Sales")
plt.show("Region")
```

### Business Impact

This analysis helps management understand regional sales performance and identify high-performing and underperforming regions for strategic planning and resource allocation.

## Analysis 3 — Monthly Sales Trend
### Business Question:

Which month generated the highest total sales?

```python
master_df["Month"] = master_df["Order_Date"].dt.month_name()

month_order = [
    "January",
    "February",
    "March",
    "April",
    "May",
    "June",
    "July",
    "August",
    "September",
    "October",
    "November",
    "December"
]

master_df["Month"] = pd.Categorical(
    master_df["Month"],
    categories=month_order,
    ordered=True
)

monthly_sales = (
    master_df
    .groupby("Month")
    .agg(
        Total_Sales=("Sales", "sum")
    )
    .sort_index()
)

monthly_sales
```

```python
monthly_sales.plot(
    kind="line",
    y="Total_Sales",
    marker="o",
    figsize=(10,5),
    title="Monthly Sales Trend"
)

plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)

plt.show()
```

#### Business Impact

This analysis helps management identify seasonal sales patterns, improve demand forecasting, optimize inventory planning, and schedule marketing campaigns during lower-performing months.

## Analysis 4 — Top 10 Products by Sales
### Business Question:

Which products generated the highest total sales?

```python
top_products = (
    master_df
    .groupby("Product_Name")
    .agg(
        Total_Sales=("Sales", "sum")
    )
    .sort_values("Total_Sales", ascending=False)
    .head(10)
)
top_products
```

```python
top_products.plot(
    kind="barh",
    y="Total_Sales",
    figsize=(10,6),
    title="Top 10 Products by Sales"
)

plt.xlabel("Total Sales")
plt.ylabel("Product Name")
plt.show()
```

### Business Impact

This analysis helps management identify the highest-performing products, optimize inventory planning, prioritize marketing efforts, and make informed decisions about product promotions and stock allocation.
